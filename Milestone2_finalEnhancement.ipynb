```python

from random import shuffle

sum1=0
player=0
dealer=0
bet=0
bal=100

class Deck():
    numbers= ['One','Ace','Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King']
    suits= ['Hearts','Diamonds','Spades','Clubs']
    dict1={'One':1,'Two':2,'Three':3,'Four':4,'Five':5,'Six':6,'Seven':7,'Eight':8,'Nine':9,'Ten':10,'Jack':10,'Queen':10,'King':10,'Ace':11}
    
   
    def beginning(self):
        
        global play1
        while True:
            play=input("Do you want to play(y/n)?...")
            play1=play.lower()

            if play1=='y':
                d1=Deck()
                d1.bet_Choice()
                break
            elif play1=='n':
                print("See you later!")
                break
            else:
                print("Please enter either y/n")
                
    
    def deck_suffle_player(self):
        global sum1
        shuffle(Deck.numbers)
        shuffle(Deck.suits)
        print(f"{Deck.numbers[0]} of {Deck.suits[0]}")
        
        if Deck.numbers[0] in Deck.dict1.keys():
            if player<22:
                if Deck.numbers[0] != 'Ace':
                    #print("Not ace")
                    sum1=sum1+Deck.dict1[Deck.numbers[0]]
   
                elif (Deck.numbers[0] == 'Ace' and player<=10):
                    print("Inside player<=10")
                    Deck.dict1[Deck.numbers[0]]=11
                    sum1+=Deck.dict1[Deck.numbers[0]]
                elif (Deck.numbers[0] == 'Ace' and player>10):
                    print("Inside player >10")
                    Deck.dict1[Deck.numbers[0]]=1
                    sum1+=Deck.dict1[Deck.numbers[0]]
                    
                    
    def deck_suffle_dealer(self):
            global sum1
            
            shuffle(Deck.numbers)
            shuffle(Deck.suits)
            print(f"{Deck.numbers[0]} of {Deck.suits[0]}")
            
            if Deck.numbers[0] in Deck.dict1.keys():
                    if dealer<22:

                        if Deck.numbers[0] != 'Ace':
                            #print("Not ace")
                            sum1=sum1+Deck.dict1[Deck.numbers[0]]

                        elif (Deck.numbers[0] == 'Ace' and dealer<=10):
                            print("Inside dealer <=10")
                            Deck.dict1[Deck.numbers[0]]=11
                            sum1+=Deck.dict1[Deck.numbers[0]]

                        elif (Deck.numbers[0] == 'Ace' and dealer>10):
                            print("Inside dealer>10")
                            Deck.dict1[Deck.numbers[0]]=1
                            sum1+=Deck.dict1[Deck.numbers[0]]
        
    
    def initial_cards(self):
        global player
        global dealer
        global sum1
        
        dealer=0
        print("Ace can be either 1 or 11 points. Jack/Queen/King values 10 points")
        print("Dealer cards are..")
        print("one card hidden")
        d=Deck()
        d.deck_suffle_dealer()
        dealer+=sum1
        sum1=0
        print(f"dealer in the beginning {dealer}")
        
        print("Player cards are..")
        player=0
        d=Deck()
        d.deck_suffle_player()
        player+=sum1
        sum1=0
        d.deck_suffle_player()
        player+=sum1
        sum1=0
        print(f"player in the beginning {player}")
        
        d.choice_of_player()
        
    
    def bet_Choice(self):
        global bal
        global bet
        print(f"Player's balance: {bal}")
        while True:
            try:
                bet=int(input("How much you wanna bet? Available bal : 100..."))
                
            except ValueError:
                print("Please provide an integer value")
            else:
                if bet<=100:
                    print("Game started!!!")
                    d=Deck()
                    d.initial_cards()
                    break
                elif (bet>100):
                    print(f"Please bet below {bal}")
                elif bet<0:
                    print("PLease be sensible")
                  
    

    def choice_of_player(self):
        global sum1
        global player
        global bal
        global bet
        global dealer
        global play1
        while True:
            if player<=21 and play1 =='y':
                choice=input("Do you wanna hit or stay(i.e)h/s...")
                choice1=choice.lower()

                if choice1=='h':
                    d=Deck()
                    d.deck_suffle_player()
                    player+=sum1
                    sum1=0
                    print(f"player while hitting {player}")
                    continue
                elif choice1=='s':    
                    print("Player's turn ended. it's dealer's turn now")
                    sum1=0
                    d=Deck()
                    d.dealer_choice()
                    if player<=21 and dealer<=21 and play1 =='y':
                        if player>dealer:
                            print("Player won the game")
                            print(f"Player's Bank balance after victory: {bal+bet}")
                            d=Deck()
                            d.beginning()
                            break
                        elif player<dealer:
                            print("Dealer won the game")
                            print(f"Dealer's Bank balance after victory: {bet}")
                            d=Deck()
                            d.beginning()
                            break
                        elif player==dealer:
                            print("Game is draw")
                            d=Deck()
                            d.beginning()
                            break
                    
                else:
                    print("Please enter either h/s")
            elif player>21 and play1 =='y':
                print("Player Busted")
                print("Dealer won the game!")
                print(f"Dealer's Bank balance after victory: {bet}")
                d=Deck()
                d.beginning()
                break
        
                
    def dealer_choice(self):
        print("Dealer's chance...")
        global sum1
        global dealer
        global bal
        global bet
        global play1
        print(f"Dealer's card count in hand: {dealer}")
        while True:
            if dealer<=17 and play1=='y':
                
                d=Deck()
                d.deck_suffle_dealer()
                dealer+=sum1
                sum1=0
                print(f"dealer while hitting {dealer}")
                continue
            elif dealer>17 and dealer<=21 and play1=='y':
                sum1=0
                print("Dealer chose to stay here")
                break 
            elif dealer>21 and play1=='y':
                print("Dealer Busted")
                print("Player won the game!")
                print(f"Player's Bank balance after victory: {bal+bet}")
                d=Deck()
                d.beginning()
                break
   
         
print("WELCOME TO BLACKJACK")   
play1='y'

d=Deck()
d.beginning()

```
